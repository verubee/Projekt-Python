```python
from selenium.webdriver import Firefox
from selenium.webdriver.firefox.service import Service as FirefoxService

from time import sleep
import re
import csv
```

```python
service = FirefoxService('./geckodriver')
browser = Firefox(service=service)  
browser.implicitly_wait(5)

job_names = ['data analyst', 'data engineer', 'data scientist']
```

```python
def get_organization_data(job_name):
    url = 'https://nofluffjobs.com/pl/?lang=en&criteria=keyword%3D"{job_name}"'  
    browser.get(url.format(job_name=job_name))

    sleep(1)

    try:
        coockies_btn = browser.find_element('id', 'onetrust-accept-btn-handler')
        coockies_btn.click()
        sleep(1)
    except:
        print('coockies ok')

    try:
        notifications_btn = browser.find_element('css selector', '.notification__close')
        notifications_btn.click()
        sleep(1)
    except:
        print('notification ok') 

    
    while True:
        try: 
            browser.find_element('css selector', '.tw-btn-xl').click()
            sleep(1)
        except:
            break


    orgs = browser.find_elements('css selector', 'a.posting-list-item')


    results = []

    for org in orgs:
        
        sleep(1)
     
        name = org.find_element('css selector', 'h3.posting-title__position').get_attribute('innerText')
        company = org.find_element('tag name', 'h4').text
        
        technologies = org.find_element('css selector', 'nfj-posting-item-tiles').text.replace('\n', '')
        
        location = org.find_element('tag name', 'nfj-posting-item-city').text.split()[0]
        
        salary = org.find_element('css selector', 'nfj-posting-item-salary').text
        print(salary)
        if not salary == 'Check Salary Match':
            bounds = re.findall('[0-9]+', salary.replace(' ', ''))
            low_bound = bounds[0]
            high_bound = bounds[1] if len(bounds) > 1 else bounds[0]
            currency = salary.split()[-1]
        else:
            low_bound = ''
            high_bound = ''
            currency = ''
            
        record = {
            'name': name,
            'company': company,
            'technology': technologies,
            'location': location,
            'salary': {
                'low': low_bound, 
                'high': high_bound, 
                'currency': currency  

            },           
        }
        
        results.append(record) 
 

    return results
           
```

```python
final_results = []
```

```python
for job in job_names:
    final_results.append(get_organization_data(job))

browser.close()
```

```python
final_results
```

```python
csv_file_path = "job_offer.csv"

with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    writer.writerow(['Name', 'Company', 'Location', 'Technologies', 'Salary Low', 'Salary High', 'Currency'])
 
    for job in final_results:
      
        for listing  in job:
            name = listing['name']
            company = listing['company']
            location = listing['location']
            technologies = listing['technology']
            salary_low = listing['salary']['low']
            salary_high = listing['salary']['high']
            salary_currency = listing['salary']['currency']
 
            writer.writerow([name, company, location, technologies, salary_low, salary_high, salary_currency])
```
